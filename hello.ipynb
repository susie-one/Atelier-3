## Etape 1 - Validation d'une saisie

```java
import java.util.Scanner;

    Scanner clavier = new Scanner(System.in);
    
    int nombre;

do {
    System.out.print("Veuillez saisir un entier entre 2 et 15 : ");
    nombre = clavier.nextInt();

    if (nombre < 2 || nombre > 15) {
        System.out.println("Erreur : saisie non valide.");
    }

} while (nombre < 2 || nombre > 15);

System.out.println("Valeur valide : " + nombre);
```

## Etape 2 - Dessin de ligne(s)

```java
for (int i = 0; i < nombre; i++) {
    for (int j = 0; j < nombre; j++) {
        System.out.print("* ");
    }
    System.out.println();
}

System.out.println("Souhaitez-vous un carré creux ? (oui/non) : ");
String choix = clavier.next();


for (int i = 0; i < nombre; i++) {
    for (int j = 0; j < nombre; j++) {

        if (choix.equalsIgnoreCase("oui")) {
            
            if (i == 0 || i == nombre - 1 || j == 0 || j == nombre - 1) {
                System.out.print("* ");
            } else {
                System.out.print("  ");
            }
        } else {
            
            System.out.print("* ");
        }
    }
    System.out.println();
}
```

## Etape 3 - Dessin de triangle(s)

```java
for (int i = 1; i <= nombre; i++) {

    for (int j = 1; j <= nombre - i; j++) {
        System.out.print(" ");
    }

    for (int j = 1; j <= i; j++) {
        System.out.print("* ");
    }

    System.out.println();
}

for (int taille = 3; taille <= nombre; taille++) {

    for (int i = 1; i <= taille; i++) {

        for (int j = 1; j <= nombre - i; j++) {
            System.out.print(" ");
        }

        for (int j = 1; j <= i; j++) {
            System.out.print("* ");
        }

        System.out.println();
    }
}
```

## Etape 4 - Dessin d'une maison

```java
for (int i = 1; i <= nombre; i++) {

    for (int j = 1; j <= nombre - i; j++) {
        System.out.print(" ");
    }

    for (int j = 1; j <= i; j++) {
        System.out.print("* ");
    }

    System.out.println();
}

for (int i = 1; i <= nombre - 1; i++) {
    for (int j = 1; j <= nombre; j++) {

        if (j == 1 || j == nombre) {
            System.out.print("* ");
        } else {
            System.out.print("  ");
        }
    }
    System.out.println();
}

for (int j = 1; j <= nombre; j++) {
    System.out.print("* ");
}
System.out.println();
```
